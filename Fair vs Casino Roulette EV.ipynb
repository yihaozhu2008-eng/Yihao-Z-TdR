```python
import numpy as np
import matplotlib.pyplot as plt

# PARAMETERS
n_spins = 5000                               # Nombre de tirades
bet_amount = 1                               # Import per aposta

# THEORETICAL EXPECTED VALUESimport numpy as np
import matplotlib.pyplot as plt

# PARAMETERS
n_spins = 10000              # Tirades per simulaci√≥
n_runs = 5000                # Nombre de simulacions independents
bet_amount = 1

red_numbers = {1,3,5,7,9,12,14,16,18,19,21,23,25,27,30,32,34,36}

# SIMULACIONS
final_profits = []

for _ in range(n_runs):
    spins = np.random.randint(0,37,size=n_spins)
    wins = np.array([bet_amount if s in red_numbers else -bet_amount for s in spins])
    final_profits.append(np.sum(wins))

final_profits = np.array(final_profits)

# PDF EMP√çRICA
plt.figure(figsize=(12,6))
plt.hist(final_profits, bins=60, density=True, alpha=0.7)

plt.title("Densitat de Probabilitat del Benefici Final (Casino Roulette)")
plt.xlabel("Benefici Final (‚Ç¨)")
plt.ylabel("Densitat de probabilitat")
plt.grid(True)
plt.show()

# Fair roulette ‚Üí EV = 0
ev_fair = np.zeros(n_spins)

# Casino roulette ‚Üí EV = -0.027027... per 1‚Ç¨ apostat
house_edge = -1/37                           # Equivalent a -0.027027
ev_casino = house_edge * np.arange(n_spins)

# PLOT
plt.figure(figsize=(12,6))
plt.plot(ev_fair, label="Fair Roulette (EV = 0)")
plt.plot(ev_casino, label="Casino Roulette (EV = -0.027¬∑x)")

plt.xlabel("Tirades")
plt.ylabel("Benefici Net Esperat (‚Ç¨)")
plt.grid(True)
plt.legend()
plt.show()
```
